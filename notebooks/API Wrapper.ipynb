```python
#inflation evolution on consumer prices since the pandemic in the USA :
inflation = wb.data.DataFrame('FP.CPI.TOTL.ZG', ['USA'], time=range(2020,2024))
inflation.head()
```

```python
inflation.columns = inflation.columns.str.replace('YR2020', '2020').str.replace('YR2021', '2021').str.replace('YR2022', '2022').str.replace('YR2023', '2023')
inflation

inflation_df = pd.DataFrame(inflation)
inflation_df = inflation_df.reset_index()

inflation_df = inflation_df.set_index('economy').unstack().reset_index()
inflation_df.columns = inflation_df.columns.str.replace('level_0', 'year')
inflation_df.columns = inflation_df.columns.fillna('inflation')
inflation_df
```

```python
# original csv file ('eviction_rate_data.csv') that was then cleaned in MySQL, before bringing it back here as 'rate_year.csv'
original_csv = pd.read_csv('flat_files/eviction_rate_data.csv', delimiter=';')
print(original_csv.shape)
original_csv.head()
```

```python
eviction_rate_year = pd.read_csv('rate_year.csv')
```

```python
eviction_rate_year['economy']='USA'

eviction_rate_year = eviction_rate_year.pivot(index='economy', columns='year', values='eviction_filing_rate')

eviction_rate_year_df = pd.DataFrame(eviction_rate_year)
eviction_rate_year_df = eviction_rate_year_df.reset_index()
eviction_rate_year_df = eviction_rate_year_df.set_index('economy').unstack().reset_index()
eviction_rate_year_df.columns = eviction_rate_year_df.columns.str.replace('0','eviction_rate')
eviction_rate_year_df.columns = eviction_rate_year_df.columns.fillna('eviction_rate')
eviction_rate_year_df
```

```python
eviction_rate_year = eviction_rate_year.pivot(index='economy', columns='year', values='eviction_filing_rate')

eviction_rate_year_df = pd.DataFrame(eviction_rate_year)
eviction_rate_year_df = eviction_rate_year_df.reset_index()
eviction_rate_year_df = eviction_rate_year_df.set_index('economy').unstack().reset_index()
eviction_rate_year_df.columns = eviction_rate_year_df.columns.str.replace('0','eviction_rate')
eviction_rate_year_df.columns = eviction_rate_year_df.columns.fillna('eviction_rate')
eviction_rate_year_df
```

```python
eviction_rate_year_df.dtypes
```

```python
inflation_df.dtypes
```

```python
inflation_df.year = inflation_df.year.astype('int64')
```

```python
df = pd.merge(inflation_df, eviction_rate_year_df, on=['year', 'economy'], how='outer', suffixes=('inflation', 'eviction_rate'))
df.inflation=round(df.inflation,2)
df = df.drop(df.index[-1])
df
```

```python
df1 = df.copy()
df1['eviction_rate'] = df1['eviction_rate']*100
df1
```
