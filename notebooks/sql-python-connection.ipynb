```python
import pandas as pd
from sqlalchemy import create_engine ,text
import pymysql.cursors
import os
import getpass
import urllib.parse
```

```python
pw_raw = os.environ.get('SQL')
```

```python
pw = urllib.parse.quote_plus(pw_raw)
```

```python
# engine = sqlalchemy.create_engine(f"mysql://{username}:{password}@{host}:{port}")

connection_string = 'mysql+pymysql://root:' + pw + '@localhost:3306/'
engine = create_engine(connection_string)
engine
```

```python
with engine.connect() as conn:
    conn.execute(text(f"CREATE DATABASE IF NOT EXISTS paws"))
```

```python
df_shelter= pd.read_csv('df_shelter.csv')
df_shelter
```

```python
df_shelter.to_sql('shelter_in_out',engine,'paws', if_exists='replace', index=False)
```
